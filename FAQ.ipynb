# 500: Internal Server Error with Webpdf
If you see an error when trying to make a Webpdf that looks like this:




You need to go to the Launcher and open a terminal. That will automatically install the  missing Chromium package, and then you should be able to make a Webpdf.



# Files don't seem up to date

If your files don't seem to match the ones Prof. Kitchin uses in class, it is possible you did not click on the link in Canvas to access them. This is the preferred way to access the notes, as it tries to update your files before opening them.

However, if you have changed the files, it is also possible they won't get updated.
