```python
from jupytercards import display_flashcards
import markdown

with open('flashcards.md') as f:
    txt = f.read()
    cards = txt.split('===')
    
flashcards = []

for card in cards:
    front, back = card.split('---')    
    flashcards += [{'front': markdown.markdown(front.strip()),
                    'back': markdown.markdown(back.strip())}]



display_flashcards(flashcards)

```
