# Multivariate interpolation

```python
import numpy as np

# level of saturation
X1 = np.array([38, 41, 34, 35, 31, 34, 29, 32])

# level of transisomers
X2 = np.array([47.5, 21.3, 36.5, 18.0, 29.5, 14.2, 21.0, 10.0])

# the yield
Y = np.array([66.0, 43.0,36.0, 23.0, 22.0, 14.0, 12.0, 7.6])

```
